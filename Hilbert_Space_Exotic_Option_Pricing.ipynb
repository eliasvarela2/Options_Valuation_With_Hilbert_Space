$$ \text{What's a Hilbert Space?}$$

A **Hilbert space** $\mathcal{H}$ is a vector space over $\mathbb{R}$ or $\mathbb{C}$ equipped with:

1. An inner product $\langle \cdot, \cdot \rangle$,

2. Such that it is complete with respect to the norm induced by the inner product:
   $$
   \|x\| := \sqrt{\langle x, x \rangle}
   $$

3. And it satisfies the following properties:
   - **Linearity in the first argument**: $\langle ax + by, z \rangle = a\langle x, z \rangle + b\langle y, z \rangle$ for all scalars $a, b$
   - **Conjugate symmetry**: $\langle x, y \rangle = \overline{\langle y, x \rangle}$
   - **Positivity**: $\langle x, x \rangle \geq 0$ and $\langle x, x \rangle = 0$ if and only if $x = 0$

**Classical example**

Let $\mathcal{H} = L^2([a,b])$, the space of real (or complex) functions $f : [a,b] \rightarrow \mathbb{R}$ (or $\mathbb{C}$) such that:
$$
\int_{a}^{b} |f(x)|^2 \, dx < \infty
$$

This space is equipped with the inner product:
$$
\langle f, g \rangle = \int_{a}^{b} f(x)g(x) \, dx
$$

In the complex case, the inner product is typically defined as:
$$
\langle f, g \rangle = \int_{a}^{b} f(x) \overline{g(x)} \, dx
$$

**Intuition**

A Hilbert space can be thought of as an *infinite-dimensional Euclidean space*:

- It contains vectors, which can be functions  
- Vectors can be projected onto each other  
- It has notions of orthogonality, bases, angles, and distances  
- Functional calculus can be performed on it

### What's a Vector Space?

A **vector space** $V$ over $\mathbb{R}$ or $\mathbb{C}$ is a set equipped with two operations:

- **Vector addition**: $u + v \in V$  
- **Scalar multiplication**: $\alpha v \in V$

These operations must satisfy the following axioms:

1. **Associativity of addition**:  
   $$(u + v) + w = u + (v + w)$$

2. **Commutativity of addition**:  
   $$u + v = v + u$$

3. **Existence of a zero vector**:  
   There exists $0 \in V$ such that:  
   $$v + 0 = v$$

4. **Existence of additive inverse**:  
   For every $v \in V$, there exists $-v \in V$ such that:  
   $$v + (-v) = 0$$

5. **Distributivity of scalar over vector addition**:  
   $$\alpha(u + v) = \alpha u + \alpha v$$

6. **Distributivity of vector over scalar addition**:  
   $$(\alpha + \beta)v = \alpha v + \beta v$$

7. **Associativity of scalar multiplication**:  
   $$\alpha(\beta v) = (\alpha\beta)v$$

8. **Existence of multiplicative identity**:  
   $$1 \cdot v = v$$

---

### Examples of Vector Spaces

- $\mathbb{R}^n$: column vectors of real numbers  
- $C([a,b])$: continuous functions on the interval $[a,b]$  
- $L^2([a,b])$: square-integrable functions on $[a,b]$

### Completeness → The Core of Hilbert Spaces

A **Hilbert space** is a vector space with an inner product that is **complete** with respect to the norm induced by that inner product.

---

### What is Completeness?

A space is **complete** if every Cauchy sequence has a limit **within** the space.

---

### What is a Cauchy Sequence?

A sequence $\{x_n\}$ is **Cauchy** if, for every $\epsilon > 0$, there exists an $N$ such that:

$$
\|x_n - x_m\| < \epsilon \quad \text{for all } n, m > N
$$

---

### Why Does It Matter?

- It guarantees that the space has no "holes"
- You can safely work with limits, approximations, and series of functions
- It enables the use of powerful tools such as:
  - The **orthogonal projection theorem**
  - **Orthogonal series** (e.g., Fourier bases)
  - **Existence of solutions** to functional equations


### What Are $x_n$ and $x_m$ in a Cauchy Sequence?

Both $x_n$ and $x_m$ are individual elements in a sequence $\{x_k\}_{k=1}^{\infty}$, and each of them is a **vector in the Hilbert space**.

They are **not components** of a single vector $X$, but rather **complete vectors** themselves. Notation can be confusing at first, but:

$$
x_n, x_m \in \mathcal{H}
$$

where $\mathcal{H}$ is a Hilbert space, such as $\mathbb{R}^n$, $L^2([a,b])$, or the space of square-integrable random variables.

---

### Interpretation

- The sequence $\{x_n\}$ is an **ordered set of vectors**:
  $$
  x_1, x_2, x_3, \dots \in \mathcal{H}
  $$

- What is being analyzed is whether, as $n, m \to \infty$, the **distance** between $x_n$ and $x_m$ becomes arbitrarily small.

Formally:
$$
\forall \epsilon > 0, \exists N \in \mathbb{N} \text{ such that } \|x_n - x_m\| < \epsilon \quad \forall n, m > N
$$

This means:
> Beyond a certain point in the sequence, all the vectors are very close to each other. The sequence has **stabilized**, even if it hasn't reached its limit yet.

---

### Example in $\mathbb{R}^2$

Let:
$$
x_n = \left(1 + \frac{1}{n}, \frac{1}{n} \right)
$$

Then:
- $x_n \to (1, 0)$ as $n \to \infty$
- The sequence is Cauchy because:

$$
\|x_n - x_m\| = \sqrt{\left( \frac{1}{n} - \frac{1}{m} \right)^2 + \left( \frac{1}{n} - \frac{1}{m} \right)^2}
= \sqrt{2} \cdot \left| \frac{1}{n} - \frac{1}{m} \right| \to 0 \quad \text{as } n,m \to \infty
$$


### Why Are $\mathbb{R}^n$ and $L^2([a,b])$ Hilbert Spaces?

In short: because they are **vector spaces** equipped with a **well-defined inner product**, and they are **complete** with respect to the **norm induced by that inner product**.

---

### Case 1: $\mathbb{R}^n$

$$
\mathbb{R}^n = \{ (x_1, x_2, \dots, x_n) \mid x_i \in \mathbb{R} \}
$$

Properties:

1. **Vector space**: vector addition and scalar multiplication are well defined.  
2. **Inner product** (standard Euclidean):
   $$
   \langle x, y \rangle = \sum_{i=1}^{n} x_i y_i
   $$
3. **Induced norm**:
   $$
   \|x\| = \sqrt{ \sum_{i=1}^{n} x_i^2 }
   $$
4. **Completeness**: every Cauchy sequence in $\mathbb{R}^n$ converges to a vector in $\mathbb{R}^n$.

Therefore, $\mathbb{R}^n$ satisfies all the properties of a **Hilbert space**.

### Case 2: $L^2([a,b])$

$L^2([a,b])$ is the set of real or complex functions $f : [a,b] \rightarrow \mathbb{R}$ or $\mathbb{C}$ such that:

$$
\int_{a}^{b} |f(x)|^2 \, dx < \infty
$$

In other words, it is the space of **square-integrable functions** over the interval $[a,b]$.

---

### Why Is $L^2([a,b])$ a Hilbert Space?

1. **Vector space**: the sum of functions and scalar multiplication are closed in $L^2$.  
2. **Inner product**:
   $$
   \langle f, g \rangle = \int_{a}^{b} f(x)\overline{g(x)} \, dx
   $$
3. **Induced norm**:
   $$
   \|f\| = \left( \int_{a}^{b} |f(x)|^2 \, dx \right)^{1/2}
   $$
4. **Completeness**: if $\{f_n\}$ is a Cauchy sequence in $L^2([a,b])$, then there exists a function $f \in L^2([a,b])$ such that:
   $$
   \|f_n - f\| \rightarrow 0
   $$

---

Therefore, $L^2([a,b])$ is an **infinite-dimensional Hilbert space**.

### How Can Hilbert Spaces Be Used in the Probabilistic Pricing of Options?

---

### What Do We Want to Calculate?

Let’s start by defining the problem. Given a financial derivative with payoff $f(S_T)$, we want to determine its **fair value today**, denoted $V_0$.

Under the **risk-neutral valuation framework**, the price is given by:

$$
V_0 = e^{-rT} \cdot \mathbb{E}^{\mathbb{Q}}[f(S_T)]
$$

Where:

- $r$: risk-free interest rate  
- $T$: time to maturity  
- $\mathbb{Q}$: risk-neutral probability measure  
- $f(S_T)$: function defining the payoff at maturity  

We now want to **reinterpret the expectation** $\mathbb{E}^{\mathbb{Q}}[f(S_T)]$ as an **inner product** in a functional space.

---

### The Hilbert Space Involved

Let:
$$
\mathcal{H} = L^2(\Omega, \mathcal{F}, \mathbb{Q})
$$

This is the space of **square-integrable random variables** under the risk-neutral measure $\mathbb{Q}$, where:

- Each $X \in \mathcal{H}$ is a random variable (e.g., a payoff)
- The inner product is defined as:
  $$
  \langle X, Y \rangle = \mathbb{E}^{\mathbb{Q}}[XY]
  $$

---

### Reinterpretation of the Option Price

If we assume that the constant function $1 \in \mathcal{H}$, then we can write:

$$
\mathbb{E}^{\mathbb{Q}}[f(S_T)] = \langle f(S_T), 1 \rangle
$$

So the fair value becomes:

$$
V_0 = e^{-rT} \cdot \langle f(S_T), 1 \rangle
$$

---

### Geometric Interpretation

In a Hilbert space, we can interpret random variables (or functions of $S_T$) as **vectors**.

- The inner product $\langle f(S_T), 1 \rangle$ measures the **component of the vector $f(S_T)$ along the constant function $1$**
- This component is precisely the **expected value** under $\mathbb{Q}$
- In other words, we are "projecting" $f(S_T)$ onto the constant function $1$

---

### Why Is This Useful?

This Hilbert space perspective allows us to reinterpret fundamental concepts in option pricing:

- **Pricing** as an inner product (expectation)
- **Hedging** as an **orthogonal projection** onto subspaces generated by tradable assets
- **Replication error** as the **distance in $L^2$** between the payoff and its approximation

This functional-analytic view is the basis of advanced pricing methods, such as those used in:

- The **martingale representation theorem**
- **Delta-hedging** and variance-minimizing strategies
- **Incomplete markets** and optimal replication

### What Is an Orthogonal Sequence?

An **orthogonal sequence** is an expansion of a function in terms of an orthonormal basis in a Hilbert space.

Let:
- $\mathcal{H} = L^2([a,b])$ be a Hilbert space
- $\{\phi_n\}_{n=1}^{\infty}$ be an **orthonormal basis** of $\mathcal{H}$ (e.g., sine and cosine functions)

Then, any function $f \in L^2([a,b])$ can be represented as:

$$
f(x) = \sum_{n=1}^{\infty} a_n \phi_n(x)
$$

Where the coefficients $a_n$ are given by the inner product:

$$
a_n = \langle f, \phi_n \rangle = \int_{a}^{b} f(x)\phi_n(x) \, dx
$$

This is known as a **functional decomposition** in the Hilbert space.

---

### What Does This Mean in Finance?

If $f(S_T)$ is the **payoff** of an option and we treat it as a function in $L^2$, we can write:

$$
f(S_T) = \sum_{n=1}^{\infty} a_n \phi_n(S_T)
$$

This decomposition allows us to:

1. **Approximate** $f(S_T)$ using a finite number of terms (truncate the series)
2. **Analyze** it more easily since we know the properties of each $\phi_n$
3. **Integrate** it efficiently if we know $\mathbb{E}^{\mathbb{Q}}[\phi_n(S_T)]$

---

### Connection to Option Pricing

Recall the pricing formula under risk-neutral valuation:

$$
V_0 = e^{-rT} \cdot \mathbb{E}^{\mathbb{Q}}[f(S_T)]
$$

Substituting the orthogonal expansion of $f(S_T)$:

$$
V_0 = e^{-rT} \cdot \mathbb{E}^{\mathbb{Q}}\left[ \sum_{n=1}^{\infty} a_n \phi_n(S_T) \right]
$$

By linearity of expectation:

$$
V_0 = e^{-rT} \cdot \sum_{n=1}^{\infty} a_n \cdot \mathbb{E}^{\mathbb{Q}}[\phi_n(S_T)]
$$

---

### Why Is This Powerful?

This approach transforms **pricing a complex payoff** into computing **expectations of known basis functions**, making the problem:

- More **tractable** computationally  
- More **interpretable** analytically  
- More **modular** for approximation and numerical methods

### Valuation of a Digital European Option using Hilbert Space Theory

We now move on to the valuation of a **digital European option** using the framework of **Hilbert spaces** and a **square-integrable probability space**.

In the following code cell, we will define the relevant parameters:

- $S_0$: initial asset price  
- $K$: strike price  
- $r$: risk-free interest rate  
- $\sigma$: volatility  
- $T$: time to maturity  
- $n_{\text{sim}}$: number of Monte Carlo simulations

````python
import numpy as np
import matplotlib.pyplot as plt

# === Establecemos los parámetros ===
S0 = 100               # Precio actual
K = 105                # Strike de la opción digital
r = 0.05               # Tasa libre de riesgo
sigma = 0.25           # Volatilidad
T = 0.25               # Tiempo (en años)
n_sim = 100_000        # Número de simulaciones
````

### Risk-Neutral Dynamics of the Underlying Asset

Given $S_t$, the price of the underlying asset at time $t$, we assume that under the **risk-neutral measure** $\mathbb{Q}$, its dynamics follow a Black-Scholes-type stochastic differential equation (SDE):

$$
dS_t = r S_t \, dt + \sigma S_t \, dW_t^{\mathbb{Q}}
$$

Where:
- $r \in \mathbb{R}$ is the constant **risk-free interest rate**
- $\sigma > 0$ is the **volatility** of the asset
- $W_t^{\mathbb{Q}}$ is a **Brownian motion** under the measure $\mathbb{Q}$
- $S_0$ is the **initial asset price**

---

### Analytical Solution of the SDE

The solution to this stochastic differential equation at time $T$ is:

$$
S_T = S_0 \cdot \exp\left[\left(r - \frac{1}{2} \sigma^2\right)T + \sigma \sqrt{T} \cdot Z\right], \quad Z \sim \mathcal{N}(0,1)
$$

This expression will be the basis for our Monte Carlo simulations under the risk-neutral measure $\mathbb{Q}$.

### Nexus with the Hilbert Space Perspective

In the functional framework, all of this allows us to view $S_T$ as a **random variable**:

$$
S_T : \Omega \rightarrow \mathbb{R}
$$

whose distribution under the **risk-neutral measure** $\mathbb{Q}$ enables us to compute expectations such as:

$$
\mathbb{E}^{\mathbb{Q}}[f(S_T)] = \int_{\Omega} f(S_T(\omega)) \, d\mathbb{Q}(\omega)
$$

---

### The Functional Translation

The ultimate goal is to **translate this expectation into a Hilbert space setting**, specifically within $L^2([0,1])$, by performing a **change of variables**.

Once we simulate $S_T \sim \mathbb{Q}$, we can **rescale** it into the interval $[0,1]$ via:

$$
x = \frac{S_T - a}{b - a}
$$

where $a$ and $b$ are bounds chosen to contain the support of $S_T$ with high probability.

This transformation allows us to treat $x$ as a variable in the Hilbert space $L^2([0,1])$, where the **payoff function** $f(x)$ can be approximated by an **orthonormal basis expansion**, such as the **Fourier basis**:

$$
f(x) \approx \sum_{n=1}^N a_n \phi_n(x)
$$

This perspective connects **probabilistic valuation** with **functional analysis** and enables efficient approximation and interpretation of option prices.

````python
# === Simulación bajo la medida neutral al riesgo ===
np.random.seed(42)
Z = np.random.normal(0, 1, n_sim)
S_T = S0 * np.exp((r - 0.5 * sigma**2) * T + sigma * np.sqrt(T) * Z)

# === Visualización de la distribución simulada === 
plt.figure(figsize=(10,5))
plt.hist(S_T, bins=100, density=True, color="skyblue", edgecolor="black")
plt.axvline(K, color="red", linestyle="--", label="Strike (K)")
plt.title("Distribución simulada de $S_T$ bajo $\\mathbb{Q}$")
plt.xlabel("$S_T$")
plt.ylabel("Densidad estimada")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
````

````python
# === Elegimos el intervalo [a, b] que contiene casi toda la masa de probabilidad de S_T ===
a = np.percentile(S_T, 5)
b = np.percentile(S_T, 95)

# === Reescalar S_T a x ∈ [0, 1] ===
x = (S_T - a) / (b - a)

# Aseguramos que esté dentro del intervalo (por redondeos numéricos)
x = np.clip(x, 0, 1)

# === Visualizamos la transformación ===
plt.figure(figsize=(12,6))
plt.hist(x, bins=100, color="mediumseagreen", edgecolor="black", density=True)
plt.title("Distribución de $x = \\frac{S_T - a}{b - a}$ en [0, 1]")
plt.xlabel("x")
plt.ylabel("Densidad")
plt.grid(True)
plt.tight_layout()
plt.show()
````

### Change of Variable and Rescaling to $[0,1]$

We aim to work in the Hilbert space:

$$
L^2([0,1], \mathbb{R}) = \left\{ f : [0,1] \rightarrow \mathbb{R} \ \bigg| \ \int_0^1 |f(x)|^2 dx < \infty \right\}
$$

To do this, we apply a **change of variable** to map the simulated values of $S_T$ (which live in a subset of $\mathbb{R}^+$) into the unit interval $[0,1]$.

---

### Rescaling Transformation

Let:
- $a = \text{5th percentile of } S_T$
- $b = \text{95th percentile of } S_T$

We define the rescaled variable:

$$
x = \frac{S_T - a}{b - a}
$$

This maps:
- $S_T = a$ to $x = 0$
- $S_T = b$ to $x = 1$

Thus, $x \in [0,1]$, making it compatible with basis functions defined on $L^2([0,1])$.

---

### Rewriting the Payoff Function

We now express the **digital option payoff** in terms of the rescaled variable $x$.

Let:

$$
\kappa := \frac{K - a}{b - a}
$$

Then the payoff becomes:

$$
f(x) = \mathbb{1}_{x > \kappa}
$$

This is a **step function** in $[0,1]$, which we can now approximate using an orthonormal basis in $L^2([0,1])$ (e.g., Fourier series).

This transformation is essential to apply Hilbert space methods for pricing and approximation in a functional framework.

### Digital Cash-or-Nothing European Call

A **digital cash-or-nothing European call option** has the following payoff at maturity $T$:

$$
f(S_T) = \mathbb{1}_{\{S_T > K\}}
$$

That is, it pays **1 unit of currency** if the asset price at maturity exceeds the strike price $K$, and **0 otherwise**.

---

### Risk-Neutral Valuation

The fair price at time $t = 0$ is given by the **discounted expected value** under the risk-neutral measure $\mathbb{Q}$:

$$
V_0 = e^{-rT} \cdot \mathbb{E}^{\mathbb{Q}}[f(S_T)]
= e^{-rT} \cdot \int_{\Omega} \mathbb{1}_{\{S_T(\omega) > K\}} \, d\mathbb{Q}(\omega)
$$

This expectation is simply the **risk-neutral probability** that $S_T > K$:

$$
V_0 = e^{-rT} \cdot \mathbb{Q}(S_T > K)
$$

---

### Closed-Form Solution (Black-Scholes)

Under the Black-Scholes model, this probability has a closed-form expression involving the standard normal CDF $\Phi$:

$$
V_0^{BS} = e^{-rT} \cdot \Phi(d_2)
$$

where:

$$
d_2 = \frac{\log\left(\frac{S_0}{K}\right) + \left(r - \frac{1}{2} \sigma^2\right)T}{\sigma \sqrt{T}}
$$

This formula provides the **benchmark** value against which we can compare our **Hilbert-space-based approximation** of the option price.

````python
# Umbral reescalado del strike
kappa = (K - a) / (b - a)

# definimos el payoff f(x) como una función escalón (indicadora)
f_x = np.where(x > kappa, 1.0, 0.0)

# Graficar f(x)
plt.figure(figsize=(10,4))
plt.plot(np.sort(x), f_x[np.argsort(x)], lw=1.2, color='darkblue')
plt.axvline(kappa, color='red', linestyle='--', label=f"x = κ ≈ {kappa:.2f}")
plt.title("Payoff $f(x)$ de la opción digital en $L^2([0,1])$")
plt.xlabel("$x$ (reescalado de $S_T$)")
plt.ylabel("$f(x)$")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
````

### Orthonormal Basis of $L^2([0,1])$

Let us consider the **Fourier basis** in the Hilbert space $L^2([0,1])$:

- $\phi_0(x) = 1$
- $\phi_n(x) = \sqrt{2} \cos(n \pi x)$ for $n \in \mathbb{N}$

This set of functions forms an **orthonormal basis** of $L^2([0,1])$ with the following properties:

- **Orthogonality**:
  $$
  \langle \phi_n, \phi_m \rangle = \delta_{nm}
  $$

- **Density**:
  The linear span $\text{Span}\{\phi_n\}$ is **dense** in $L^2([0,1])$, meaning that **any** function $f \in L^2([0,1])$ can be represented as:

  $$
  f(x) = \sum_{n=0}^{\infty} a_n \phi_n(x)
  $$

  where the Fourier coefficients are given by:

  $$
  a_n = \langle f, \phi_n \rangle = \int_0^1 f(x) \phi_n(x) \, dx
  $$

This expansion provides a powerful way to approximate and analyze functions in $L^2([0,1])$, including option payoff functions, using only inner products with known basis functions.

### Monte Carlo Estimation of Fourier Coefficients

Since we only know the values of the function $f(x)$ at a finite number of sample points $\{x_i\}_{i=1}^M$, drawn according to the distribution induced by $\mathbb{Q}$, we **approximate the inner product** using the **empirical mean**.

The Fourier coefficient $a_n$ is defined as:

$$
a_n = \langle f, \phi_n \rangle = \int_0^1 f(x) \phi_n(x) \, dx
$$

We approximate this using the Monte Carlo estimator:

$$
a_n \approx \frac{1}{M} \sum_{i=1}^M f(x_i) \phi_n(x_i)
$$

This approximation is justified by the **Law of Large Numbers**:

$$
\mathbb{E}\left[\frac{1}{M} \sum_{i=1}^M f(x_i) \phi_n(x_i)\right] = \langle f, \phi_n \rangle
$$

Therefore, as $M \to \infty$, the empirical average converges to the true inner product.

This method allows us to numerically compute the projection of a function $f$ onto the orthonormal basis $\{\phi_n\}$ using only sample values.

````python
# === Definimos la base de Fourier ===
def phi(n, x):
    if n == 0:
        return np.ones_like(x)
    else:
        return np.sqrt(2) * np.cos(n * np.pi * x)

# === Número de términos de la expansión === 
N = 20 # Lo podemos aumentar si queremos mayor precisión

# === Calcular coeficientes a_n === 
a_n = []
for n in range(N + 1):
    a = np.mean(f_x * phi(n, x)) # producto interno empírico en L^2
    a_n.append(a)

# === Reconstrucción de f(x) con N términos === 
f_reconstruida = np.zeros_like(x)
for n in range(N + 1):
    f_reconstruida += a_n[n] * phi(n, x)
    
# === Graficar f(x) original vs. reconstruida ===
x_sorted = np.sort(x)
plt.figure(figsize=(10,5))
plt.plot(x_sorted, f_x[np.argsort(x)], label="f(x) original", color="blue", linewidth=2)
plt.plot(x_sorted, f_reconstruida[np.argsort(x)], label=f"Reconstrucción Fourier N={N}", linestyle="--", color="orange")
plt.title("Aproximación de $f(x)$ mediante base de Fourier en $L^2([0, 1])$")
plt.xlabel("$x$")
plt.ylabel("$f(x)$")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
````

### Estimation of the Expected Value $\mathbb{E}^{\mathbb{Q}}[\phi_n(x)]$

To approximate the price of the digital option via the truncated orthogonal expansion:

$$
\mathbb{E}^{\mathbb{Q}}[f(x)] \approx \sum_{n=0}^{N} a_n \cdot \mathbb{E}^{\mathbb{Q}}[\phi_n(x)]
$$

we also need to estimate the expected value of each basis function $\phi_n$ under the measure $\mathbb{Q}$.

---

### Monte Carlo Estimator

Just as we approximated the inner product using sample points $\{x_i\}_{i=1}^M$, we can estimate:

$$
\mathbb{E}^{\mathbb{Q}}[\phi_n(x)] \approx \frac{1}{M} \sum_{i=1}^{M} \phi_n(x_i)
$$

This is valid due to the **Law of Large Numbers**, since the $x_i$ are realizations of a random variable with distribution induced by $\mathbb{Q}$ over $[0,1]$.

Combining this with the previous approximation of $a_n$, we obtain a **fully empirical approximation** for the expectation of $f(x)$:

$$
\mathbb{E}^{\mathbb{Q}}[f(x)] \approx \sum_{n=0}^{N} \left( \frac{1}{M} \sum_{i=1}^{M} f(x_i)\phi_n(x_i) \right) \cdot \left( \frac{1}{M} \sum_{i=1}^{M} \phi_n(x_i) \right)
$$

This formulation allows us to **estimate option prices** using only function evaluations and basis projections, within the Hilbert space framework.

### Functional Approximation of the Option Price

Finally, we assemble the **functional approximation** of the option price using the truncated Fourier expansion.

Let:

$$
V_0^{(N)} := e^{-rT} \cdot \sum_{n=0}^{N} a_n \cdot \mathbb{E}^{\mathbb{Q}}[\phi_n(x)]
$$

This is our estimated price of the option based on the first $N$ terms of the orthonormal basis expansion.

---

### Convergence

Under mild regularity conditions on the payoff function $f(x)$, this approximation satisfies:

$$
\lim_{N \to \infty} V_0^{(N)} = V_0
$$

That is, the approximation **converges** to the true price of the option as $N$ increases.

However, in practice, we often find that only a **small number of terms** is sufficient to obtain an accurate estimate of the price — especially when the basis functions are well-suited to the shape of the payoff.

This is the power of the **Hilbert space formulation**: it transforms option pricing into a problem of **functional approximation**.

````python
# === Estimar E[phi_n(x)] usando Monte Carlo ===
E_phi_n = []
for n in range(N + 1):
    E_phi_n.append(np.mean(phi(n, x))) # esperanza empírica

# === Ensamblar el valor funcional V_0 ===
a_n = np.array(a_n)
E_phi_n = np.array(E_phi_n)

V_0 = np.exp(-r * T) * np.sum(a_n * E_phi_n)

# === Comparamos con el valor exacto de Black-Scholes para una opción digital ===
from scipy.stats import norm
d2 = (np.log(S0 / K) + (r - 0.5 * sigma**2) * T) / (sigma * np.sqrt(T))
V_BS = np.exp(-r * T) * norm.cdf(d2)

# === Mostrar resultados ===
print(f"Precio funcional aproximado (V_0): {V_0:.6f}")
print(f"Precio exacto Black-Scholes: {V_BS:.6f}")
print(f"Error V_0 respecto a V_BS: {(V_0 - V_BS):.6f}")
````

### Fejér (Cesàro) Smoothing of the Fourier Series

Since the payoff function $f(x)$ is **discontinuous**, its Fourier series **converges in the $L^2$ sense** (quadratic mean), but **not pointwise**. This is due to the **Gibbs phenomenon**, where oscillations near discontinuities persist even as $N \to \infty$.

To mitigate this, we use the **Fejér (Cesàro) smoothing** technique.

---

### Smoothed Approximation

The **Fejér-smoothed Fourier approximation** of the function is given by:

$$
f^{\text{Fejér}}(x) := \sum_{n=0}^{N} \left(1 - \frac{n}{N + 1} \right) a_n \phi_n(x)
$$

This is equivalent to averaging the partial sums of the Fourier series:

$$
f^{\text{Fejér}}(x) = \frac{1}{N + 1} \sum_{k=0}^{N} \left( \sum_{n=0}^{k} a_n \phi_n(x) \right)
$$

Fejér’s theorem ensures that this approximation **converges uniformly** (in the sense of Cesàro means) even for discontinuous functions.

---

### Functional Option Price with Fejér Smoothing

In the context of option pricing, we can apply Fejér smoothing to the functional approximation as follows:

$$
V_0^{\text{Fejér}} := e^{-rT} \cdot \sum_{n=0}^{N} \left(1 - \frac{n}{N + 1} \right) a_n \cdot \mathbb{E}^{\mathbb{Q}}[\phi_n(x)]
$$

This version often provides a **more accurate and stable estimate** for options with discontinuous payoffs, such as digital options.

The smoother convergence of the Fejér approximation makes it particularly useful when truncating the series at a relatively small $N$.

````python
import matplotlib.pyplot as plt

# === Evaluar error para múltiples valores de N ===
Ns = np.arange(1, 501, 10)
errores_fourier = []
errores_fejer = []

for N in Ns:
    a_n = []
    E_phi_n = []
    for n in range(N + 1):
        phi_n_x = phi(n, x)
        a = np.mean(f_x * phi_n_x)
        e = np.mean(phi_n_x)
        a_n.append(a)
        E_phi_n.append(e)

    # Precio directo
    V_directo = np.exp(-r * T) * np.sum(np.array(a_n) * np.array(E_phi_n))
    errores_fourier.append(abs(V_directo - V_BS))

    # Precio suavizado (Fejér)
    V_fejer = np.exp(-r * T) * np.sum([(1 - n / (N + 1)) * a_n[n] * E_phi_n[n] for n in range(N + 1)])
    errores_fejer.append(abs(V_fejer - V_BS))

# === Graficar errores ===
plt.figure(figsize=(10, 6))
plt.plot(Ns, errores_fourier, label="Error Fourier directo", linewidth=2, color="orange")
plt.plot(Ns, errores_fejer, label="Error Fejér (suavizado)", linewidth=2, color="green")
plt.axhline(y=0, color='black', linestyle='--')
plt.title("Error absoluto vs. número de términos N")
plt.xlabel("Número de términos N")
plt.ylabel("Error absoluto |V₀ - V_BS|")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
````

### From Black-Scholes to Hilbert-Space-Based Pricing of Exotic Options

Now that we've built intuition in a context where Black-Scholes is applicable — specifically, the pricing of a digital European option — and verified that our **Hilbert space approach with orthogonal function expansions** converges to the same result, we are ready to **extend this method to exotic derivatives** where Black-Scholes no longer yields a closed-form solution.

A prime example is the **Asian option**, which depends on the average value of the underlying asset over time.

$$\overline{S} = \frac{1}{N}\sum_{i=1}^{N}S_{t_{i}}$$

---

### Step 1: Risk-Neutral Dynamics of the Asset

Under the **risk-neutral probability measure** $\mathbb{Q}$, the asset price $S_t$ follows the stochastic differential equation (SDE):

$$
dS_t = r S_t \, dt + \sigma S_t \, dW_t^{\mathbb{Q}}
$$

Where:
- $r$ is the risk-free interest rate
- $\sigma$ is the volatility of the asset
- $W_t^{\mathbb{Q}}$ is a standard Brownian motion under $\mathbb{Q}$

The **analytical solution** to this SDE (via Itô calculus) is:

$$
S_t = S_0 \cdot \exp\left( \left(r - \frac{1}{2}\sigma^2\right)t + \sigma W_t^{\mathbb{Q}} \right)
$$

---

### Step 2: Simulating the Process

To simulate paths of $S_t$ from $t = 0$ to $t = T$, we divide the time interval into $n$ equal steps:

$$
\Delta t = \frac{T}{n}
$$

We generate the asset values $S_{t_1}, S_{t_2}, \dots, S_{t_n}$ **iteratively**. On the logarithmic scale, we use:

$$
\log S_{t_{i+1}} = \log S_{t_i} + \left(r - \frac{1}{2}\sigma^2\right) \Delta t + \sigma \sqrt{\Delta t} \cdot Z_i
$$

where $Z_i \sim \mathcal{N}(0,1)$ are independent standard normal variables.

---

### Step 3: Computational Simulation

We define key simulation parameters:

```python
S0 = 100            # Initial price
r = 0.05            # Risk-free rate
sigma = 0.2         # Volatility
T = 1.0             # Time to maturity (in years)
n_steps = 100       # Number of time steps
n_sim = 100_000     # Number of simulated paths
dt = T / n_steps    # Time step size

We simulate the Brownian motion increments:
    
Z = np.random.normal(0, 1, size=(n_sim, n_steps))
increments = (r - 0.5 * sigma**2) * dt + sigma * np.sqrt(dt) * Z


This corresponds to:
$$ \Delta logS_{t} = (r - \frac{1}{2}\sigma^{2})\Delta t + \sigma\sqrt{\delta t}\cdot Z_{i} $$

````python
import numpy as np
import matplotlib.pyplot as plt

# === Parámetros del modelo ===
S0 = 100
r = 0.05
sigma = 0.6
T = 1.0
n_steps = 100  # número de pasos de tiempo
n_sim = 100_000  # número de trayectorias

dt = T / n_steps
np.random.seed(42)
Z = np.random.normal(0, 1, size=(n_sim, n_steps))
increments = (r - 0.5 * sigma**2) * dt + sigma * np.sqrt(dt) * Z

# Simulación log-normal (Euler)
log_paths = np.cumsum(increments, axis=1)
S_paths = S0 * np.exp(log_paths)

# Agregamos S0 al inicio para completar la trayectoria
S_paths = np.hstack((S0 * np.ones((n_sim, 1)), S_paths))

# === Visualizamos algunas trayectorias ===
plt.figure(figsize=(12,6))
for i in range(10):
    plt.plot(np.linspace(0, T, n_steps + 1), S_paths[i, :], lw=1)
plt.title("Trayectorias simuladas de $S_t$ bajo $\\mathbb{Q}$")
plt.xlabel("Tiempo")
plt.ylabel("Precio del subyacente")
plt.grid(True)
plt.tight_layout()
plt.show()
````

### Payoff of an Asian Call Option and Functional Transformation

We now define the **payoff of an Asian call option**, which depends on the **arithmetic average** of the underlying asset over multiple time points $t_1, \dots, t_n$:

$$
\text{Payoff} = \max(\overline{S} - K, 0), \quad \text{where } \overline{S} = \frac{1}{n} \sum_{i=1}^{n} S_{t_i}
$$

In order to apply **orthogonal functional techniques**, we rescale $\overline{S}$ to the interval $[0,1]$, so it can be treated as a function $f(x) \in L^2([0,1])$.

---

### Definition of the Average

Let $S_t^{(j)}$ be the $j$-th simulated trajectory. We compute the arithmetic average for each simulation:

$$
\overline{S}^{(j)} := \frac{1}{n} \sum_{i=1}^{n} S_{t_i}^{(j)}
$$

for $j = 1, \dots, N_{\text{sim}}$. This gives us a sample $\{\overline{S}^{(j)}\}_{j=1}^{N_{\text{sim}}}$ of average prices.

---

### Payoff Function of the Asian Option

The Asian option payoff for each trajectory is given by:

$$
f(\overline{S}) = \max(\overline{S} - K, 0)
$$

This is the function we want to represent in the functional space $L^2([a, b])$. However, to work within $L^2([0,1])$, we must **rescale** the variable.

---

### Rescaling the Support to $[0,1]$

We choose:

- $a = \text{5th percentile of } \overline{S}$
- $b = \text{95th percentile of } \overline{S}$

Then we define the rescaled variable:

$$
x^{(j)} := \frac{\overline{S}^{(j)} - a}{b - a} \in [0,1]
$$

---

### Rescaled Payoff Function

To maintain consistency with the original payoff, we define the function $f(x^{(j)})$ as:

$$
f(x^{(j)}) = \max\left((x^{(j)}(b - a) + a) - K, \, 0\right)
$$

This is now a function defined on the compact interval $[0,1]$, and therefore belongs to the Hilbert space $L^2([0,1])$.

We can now apply the same **orthogonal projection techniques** (e.g., Fourier expansion and Fejér smoothing) to approximate the price of the Asian option.

### Fourier Approximation of the Payoff Function in $L^2([0,1])$

We want to approximate the function $f(x) \in L^2([0,1])$ using a truncated **orthonormal Fourier expansion**:

$$
f(x) \approx \sum_{n=0}^{N} a_n \, \phi_n(x)
$$

---

### Basis Functions

The functions $\{\phi_n(x)\}$ form an **orthonormal basis** of $L^2([0,1])$, specifically:

- $\phi_0(x) = 1$
- $\phi_n(x) = \sqrt{2} \cos(n \pi x)$ for all $n \geq 1$

These functions satisfy:

$$
\langle \phi_n, \phi_m \rangle = \delta_{nm}
$$

---

### Fourier Coefficients

The coefficients $a_n$ are computed via the inner product in $L^2$:

$$
a_n = \langle f, \phi_n \rangle = \int_0^1 f(x) \, \phi_n(x) \, dx
$$

Since we only know the function at discrete sample points $\{x_j\}_{j=1}^M$, we approximate the integral with the **empirical mean**:

$$
a_n \approx \frac{1}{M} \sum_{j=1}^{M} f(x_j) \, \phi_n(x_j)
$$

This Monte Carlo estimator converges to the true coefficient $a_n$ as $M \to \infty$, by the **Law of Large Numbers**.

We can now use these coefficients to reconstruct or approximate the function $f(x)$ and, consequently, estimate the price of the Asian option in a functional framework.

````python
# === Paso 2: Cálculo del promedio de cada trayectoria ===
S_promedios = np.mean(S_paths[:, 1:], axis=1)  # desde t=Δt hasta T

# === Elegimos intervalo [a, b] que contenga la mayoría de la masa ===
a = np.percentile(S_promedios, 5)
b = np.percentile(S_promedios, 95)

# === Reescalar a x ∈ [0, 1] ===
x = (S_promedios - a) / (b - a)
x = np.clip(x, 0, 1)
K = 90

# === Definimos el payoff reescalado ===
S_promedio_transformado = x * (b - a) + a
f_x = np.maximum(S_promedio_transformado - K, 0)


# === Visualización ===
import matplotlib.pyplot as plt

plt.figure(figsize=(12,5))
plt.hist(x, bins=100, color="teal", edgecolor="black", density=True)
plt.title("Distribución del promedio reescalado $x = (\\bar{S} - a)/(b - a)$")
plt.xlabel("x")
plt.ylabel("Densidad")
plt.grid(True)
plt.tight_layout()
plt.show()

plt.figure(figsize=(10,4))
plt.plot(np.sort(x), f_x[np.argsort(x)], color="darkred", lw=1.5)
plt.title("Payoff reescalado $f(x)$ de la opción asiática")
plt.xlabel("$x$")
plt.ylabel("$f(x)$")
plt.grid(True)
plt.tight_layout()
plt.show()
````

````python
# === Paso 3: Definir base ortonormal de Fourier en L^2([0,1]) ===
def phi(n, x):
    if n == 0:
        return np.ones_like(x)
    else:
        return np.sqrt(2) * np.cos(n * np.pi * x)

# === Número de términos de la expansión ===
N = 70  # puedes ajustar esto para mayor precisión

# === Calcular coeficientes a_n (producto interno empírico) ===
a_n = []
for n in range(N + 1):
    a = np.mean(f_x * phi(n, x))
    a_n.append(a)

# === Reconstrucción de f(x) usando N términos ===
f_reconstruida = np.zeros_like(x)
for n in range(N + 1):
    f_reconstruida += a_n[n] * phi(n, x)

# === Graficar f(x) original vs reconstrucción de Fourier ===
import matplotlib.pyplot as plt

x_sorted = np.sort(x)
plt.figure(figsize=(10,5))
plt.plot(x_sorted, f_x[np.argsort(x)], label="f(x) original", color="blue", linewidth=2)
plt.plot(x_sorted, f_reconstruida[np.argsort(x)], label=f"Reconstrucción Fourier N={N}", linestyle="--", color="orange")
plt.title("Aproximación de $f(x)$ con base de Fourier en $L^2([0,1])$")
plt.xlabel("$x$")
plt.ylabel("$f(x)$")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
````

### Approximation of the Present Value of the Asian Option

We aim to compute the **present value** of the option:

$$
V_0 = e^{-rT} \cdot \mathbb{E}^{\mathbb{Q}}[f(x)]
$$

---

### Functional Approximation

Given that we have already approximated the function $f(x)$ using a truncated Fourier expansion:

$$
f(x) \approx \sum_{n=0}^{N} a_n \, \phi_n(x)
$$

We can approximate the expectation of $f(x)$ under the measure $\mathbb{Q}$ as:

$$
\mathbb{E}^{\mathbb{Q}}[f(x)] \approx \sum_{n=0}^{N} a_n \cdot \mathbb{E}^{\mathbb{Q}}[\phi_n(x)]
$$

---

### Final Approximation of the Price

Substituting this into the pricing formula, we get the functional approximation of the option value:

$$
V_0 \approx e^{-rT} \cdot \sum_{n=0}^{N} a_n \cdot \mathbb{E}^{\mathbb{Q}}[\phi_n(x)]
$$

This expression allows us to compute the present value of the Asian option using the Hilbert space framework and orthonormal basis expansions. It transforms the pricing problem into the computation of inner products and expected values of known basis functions.

````python
# === Estimar E[phi_n(x)] usando Monte Carlo ===
E_phi_n = []
for n in range(N + 1):
    E_phi_n.append(np.mean(phi(n, x)))  # esperanza empírica bajo medida Q

# === Ensamblar el valor funcional V_0 ===
a_n = np.array(a_n)
E_phi_n = np.array(E_phi_n)

V_0 = np.exp(-r * T) * np.sum(a_n * E_phi_n)

# === Mostrar resultado ===
print(f"Precio funcional aproximado (V_0): {V_0:.6f}")
````

````python
prob_rentable = np.mean(S_promedios > K)
print(f"Probabilidad de que la opción sea rentable (in-the-money): {prob_rentable:.4f}")
````
